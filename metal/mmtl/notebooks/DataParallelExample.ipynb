```python
from metal.mmtl.glue.glue_tasks import create_tasks

task_names = [
    "COLA", 
    "SST2", 
    "MNLI",
    "RTE",
    "WNLI",
    "QQP",
    "MRPC",
    "STSB",
    "QNLI"
]
tasks = create_tasks(task_names,
                     max_datapoints=1000,
                     dl_kwargs={
                         "batch_size": 32
                     }
                    )
```

```python
import os
print('Cuda devices: ',os.environ['CUDA_VISIBLE_DEVICES'])
```

```python
from metal.mmtl.metal_model import MetalModel

model = MetalModel(tasks)
```

```python
from metal.mmtl.trainer import MultitaskTrainer
trainer = MultitaskTrainer()
```

```python
trainer.train_model(
    model,
    tasks,
    checkpoint_metric="model/train/loss",
    n_epochs=1,
    progress_bar=True
)
```
